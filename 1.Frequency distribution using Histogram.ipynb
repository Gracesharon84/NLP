# FREQUENCY DISTRIBUTION USING HISTOGRAMS

1.1 Simple Frequency Distribution

```python
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
text='Natural Language Processing (NLP) with Python is fun and educational. Python simplifies NLP tasks.'
tokens=tokenizer.tokenize(text)
print("Tokens:\n",tokens)
```

```python
fdist=FreqDist(tokens)
print("\n Word Frequency Distribution:")
for word,freq in fdist.items():
    print(f'{word}:{freq}')
fdist.plot(title='Simple Frequency Distribution Plot')
plt.show()
```

```python
#histogram plot
plt.hist(fdist.values(), bins=2, alpha=0.7)
plt.title('Word Frequency Distribution')
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.show()
```

1.2 Advanced Frequency Distribution with Stopword Removal

```python
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
```

```python
tokenizer=TreebankWordTokenizer()
stop_words=set(stopwords.words('english'))
```

```python
text='Natural Language Processing with Python is interesting and and useful. Python makes NLP tasks easier and more efficient'
tokens=tokenizer.tokenize(text)
```

```python
filtered_tokens=[word for word in tokens if word.lower() not in stop_words and word.isalpha()]
print("Filtered Tokens(without stopwords):\n",filtered_tokens)
```

```python
fdist=FreqDist(filtered_tokens)
print("\n Word Frequency Distribution(without stopwords):")
for word,freq in fdist.items():
    print(f"{word}:{freq}")
fdist.plot(title='Word Frequency Distribution(without stopwords)')
plt.show()
```

```python
plt.hist(fdist.values(), bins=2, alpha=0.7)
plt.title('Word Frequency Distribution (without stopwords)')
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.show()
```
